# Limpeza de Dados

## Percentual dos Valores Ausentes

```python
# Total de Valores Ausentes
total_ausentes = df.isnull().sum().sum()

# Total de Células
total_celulas = np.product(df.shape)

# Percentual dos valores ausentes
percentual_ausentes = total_ausentes/total_celulas * 100

# Print
print(f'Porcentagem dos dados ausentes: {percentual_ausentes:.2f}%')
```

## Percentual das Linhas com Valores Ausentes

```python
# Número Total de Linhas
total_linhas = df.shape[0]

# Total de Linhas Ausentes
count = 0
for idx, row in df.iterrows():
    if any(row.isna()):
        count += 1
linhas_ausentes = count

# Percentual das linhas com valores ausentes
percentual_linhas_ausentes = linhas_ausentes/total_linhas * 100

# Print
print(f'Percentual das linhas com valores ausentes: {percentual_linhas_ausentes:.2f}%')
```

## Percentual de Valores Ausentes por Colunas

```python
# Total de Valores Ausentes
total_ausentes = df.isnull().sum() 

# Porcentagem de valores ausentes em cada coluna
percentual_por_coluna = total_ausentes / len(df) * 100

# Type de cada coluna
type_coluna = df.dtypes

# Criar um dataframe com os resultados
df_dados_linhas_ausentes = pd.DataFrame({
    'Valores Ausentes':total_ausentes,
    'Percentual':percentual_por_coluna,
    'Type':type_coluna
}).sort_values('Percentual', ascending=False)

# Print
df_dados_linhas_ausentes
```

## Remover Colunas

```python
# Colunas com mais de 30% dos dados ausentes
df_dados_linhas_ausentes[df_dados_linhas_ausentes['Percentual'] > 30.00]
```

```python
# Apatgar colunas
colunas_excluir = df_dados_linhas_ausentes[df_dados_linhas_ausentes['Percentual'] > 30.00].index.to_list()
df = df.drop(columns=colunas_excluir)
```

## Alterar o Tipo de Dados

### Ano Modelo

```python
# Renomear valores NaN
df['ANO_MODELO'] = df['ANO_MODELO'].fillna(0.0)

# Mudar o tipo de Dados do Ano Modelo. float64 -> int64
df['ANO_MODELO'] = df['ANO_MODELO'].astype('int64')

# Verificar o tipo de dado
df['ANO_MODELO'].dtypes
```

### CEP

```python
# Renomear os valores NaN
df['CEP'] = df['CEP'].fillna(0000000)

# Mudar o tipo de dado do CEP. float64 -> int64 > string
df['CEP'] = df['CEP'].astype('int64').astype(str)

# Verificar o tipo de dado
df['CEP'].dtypes
```

### Número Logradouro

```python
# Renomear valores NaN
df['NUMERO_LOGRADOURO'] = df['NUMERO_LOGRADOURO'].fillna(0.0)

# Mudar o tipo de Dados do Número do Logradouro. float64 -> int64
df['NUMERO_LOGRADOURO'] = df['NUMERO_LOGRADOURO'].astype('int64')

# Verificar o tipo de dado
df['NUMERO_LOGRADOURO'].dtypes
```

# Análise Exploratória de Dados

## Roubos por Mês

```python
# Copiar o dataframe original
roubos_por_mes = df.copy()

# Extrair o mês
roubos_por_mes['W_MES_OCORRENCIA_BO'] = roubos_por_mes['DATA_OCORRENCIA_BO'].dt.month

# Renomear valores NaN
roubos_por_mes['W_MES_OCORRENCIA_BO'] = roubos_por_mes['W_MES_OCORRENCIA_BO'].fillna(0)

# Mudar o tipo de Dados do Mês da Ocorrência. float64 -> int64
roubos_por_mes['W_MES_OCORRENCIA_BO'] = roubos_por_mes['W_MES_OCORRENCIA_BO'].astype('int64')

# Excluir mês 0, pois eram dados nulos
roubos_por_mes = roubos_por_mes.loc[roubos_por_mes['W_MES_OCORRENCIA_BO'] != 0]

# print
roubos_por_mes.head()
```

```python
sns.countplot(roubos_por_mes, x="W_MES_OCORRENCIA_BO")
```

## Roubos por Dia da Semana

```python
roubos_por_dia = df.copy()
```

```python
# Extrair o dia da semana
roubos_por_dia['W_DIA_OCORRENCIA_BO'] = roubos_por_dia['DATA_OCORRENCIA_BO'].dt.day_name()

# Dicionário com os dias da semana em inglês e portugues
dias_semana = {
    'Monday': 'Segunda-feira',
    'Tuesday': 'Terça-feira',
    'Wednesday': 'Quarta-feira',
    'Thursday': 'Quinta-feira',
    'Friday': 'Sexta-feira',
    'Saturday': 'Sábado',
    'Sunday': 'Domingo'
}

# Traduzir os nomes para portugues
roubos_por_dia['W_DIA_OCORRENCIA_BO'] = roubos_por_dia['W_DIA_OCORRENCIA_BO'].map(dias_semana)

# print
roubos_por_dia.head()
```

```python
sns.countplot(roubos_por_dia, x="W_DIA_OCORRENCIA_BO")
```

## Modelos de Carros Mais Roubados

```python
# Copiar o dataframe original
roubos_modelos_carros = df.copy()

# Extrair apemas automóveis
roubos_modelos_carros = roubos_modelos_carros.loc[roubos_modelos_carros['DESCR_TIPO_VEICULO'] == 'Automovel']

# Extrair Marca e Veiculo em colunas separadas
roubos_modelos_carros[['W_MARCA', 'W_VEICULO', 'W_OUTROS']] = roubos_modelos_carros['DESCR_MARCA_VEICULO'].str.split('/', expand=True)

# Apagar coluna W_OUTROS
roubos_modelos_carros = roubos_modelos_carros.drop(['W_OUTROS'], axis=1)
```

```python
# Renomear para Chevrolet
roubos_modelos_carros['W_MARCA'] = roubos_modelos_carros['W_MARCA'].replace(['CHEVROLET', 'GM', 'CHEV'], 'Chevrolet')

# Renomear para Fiat
roubos_modelos_carros['W_MARCA'] = roubos_modelos_carros['W_MARCA'].replace(['FIAT', 'FIAT UNO 1.6'], 'Fiat')

# Renomear para VW
roubos_modelos_carros['W_MARCA'] = roubos_modelos_carros['W_MARCA'].replace(['VW', 'VOLKSWAGEM', 'VOLKSWAGEN', 'VOLKWAGEN'], 'VW')

# Renomear para Hyundai
roubos_modelos_carros['W_MARCA'] = roubos_modelos_carros['W_MARCA'].replace(['HYUNDAI', 'HYUNDAY', 'HYUNDA'], 'Hyundai')

# Renomear para Citroen
roubos_modelos_carros['W_MARCA'] = roubos_modelos_carros['W_MARCA'].replace(['CITROEN', 'CITROE', 'CITROEN PICASSO II20GLXA'], 'Citroen')
```
