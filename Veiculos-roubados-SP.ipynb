# Limpeza de Dados

## Percentual dos Valores Ausentes

```python
# Total de Valores Ausentes
total_ausentes = df.isnull().sum().sum()

# Total de Células
total_celulas = np.product(df.shape)

# Percentual dos valores ausentes
percentual_ausentes = total_ausentes/total_celulas * 100

# Print
print(f'Porcentagem dos dados ausentes: {percentual_ausentes:.2f}%')
```

## Percentual das Linhas com Valores Ausentes

```python
# Número Total de Linhas
total_linhas = df.shape[0]

# Total de Linhas Ausentes
count = 0
for idx, row in df.iterrows():
    if any(row.isna()):
        count += 1
linhas_ausentes = count

# Percentual das linhas com valores ausentes
percentual_linhas_ausentes = linhas_ausentes/total_linhas * 100

# Print
print(f'Percentual das linhas com valores ausentes: {percentual_linhas_ausentes:.2f}%')
```

## Percentual de Valores Ausentes por Colunas

```python
# Total de Valores Ausentes
total_ausentes = df.isnull().sum() 

# Porcentagem de valores ausentes em cada coluna
percentual_por_coluna = total_ausentes / len(df) * 100

# Type de cada coluna
type_coluna = df.dtypes

# Criar um dataframe com os resultados
df_dados_linhas_ausentes = pd.DataFrame({
    'Valores Ausentes':total_ausentes,
    'Percentual':percentual_por_coluna,
    'Type':type_coluna
}).sort_values('Percentual', ascending=False)

# Print
df_dados_linhas_ausentes
```

## Modelos de Carros Mais Roubados

```python
# Copiar o dataframe original
roubos_modelos_carros = df.copy()

# Extrair apemas automóveis
roubos_modelos_carros = roubos_modelos_carros.loc[roubos_modelos_carros['DESCR_TIPO_VEICULO'] == 'Automovel']
```

A '/' divide a marca e o tipo de veículo, assim podemos usar o método split para separar os dois.

```python
roubos_modelos_carros['DESCR_MARCA_VEICULO'].head()
```

```python
# Extrair Marca e Veiculo em colunas separadas
roubos_modelos_carros[['W_MARCA', 'W_VEICULO', 'W_OUTROS']] = roubos_modelos_carros['DESCR_MARCA_VEICULO'].str.split('/', expand=True)

# Apagar coluna W_OUTROS
roubos_modelos_carros = roubos_modelos_carros.drop(['W_OUTROS'], axis=1)
```

Nem todos são divididos pela '/'. Alguns têm 'I' na frente da marca e o tipo de veículo. Portanto, precisamos remover o 'I', extrair a marca antes do primeiro espaço e o tipo do veículo após o primeiro espaço.

```python
roubos_modelos_carros[['W_MARCA','W_VEICULO']].loc[roubos_modelos_carros['W_VEICULO'].str.contains('Chevrolet', case=False, na=False)].head()
```

Foi definida uma função para facilitar a remoção do que estiver antes da marca e do tipo de veículo.
Utilizaremos o exemplo da VW para explicar o funcionamento dessa função. 
A função busca as linhas que contêm 'VW', sem diferenciar maiúsculas de minúsculas e, além disso, não considera valores ausentes como correspondentes. Em seguida, realiza a divisão no primeiro espaço para extrair o nome da marca e o tipo de veículo. Por fim, atualiza os nomes no DataFrame original."

```python
def dividir_marca_veiculo(nome_da_marca):
    query = roubos_modelos_carros['W_VEICULO'].str.contains(nome_da_marca, case=False, na=False)
    separacao = roubos_modelos_carros.loc[query, 'W_VEICULO'].str.split(' ', n=1, expand=True)
    roubos_modelos_carros.loc[query, ['W_MARCA', 'W_VEICULO']] = separacao.values
    return print(nome_da_marca)
```

```python
lista_marcas = ['VW', 'HYUNDAI', 'CHEVROLET', 'FIAT', 'TOYOTA', 'FORD', 'GM','PEUGEOT', 'HONDA', 'RENAULT', 'CHEV', 'JEEP', 'NISSAN',
                'CITROEN', 'BMW', 'M.BENZ', 'CHERY', 'VOLKSWAGEN', 'AUDI', 'SUZUKI', 'IVECO', 'DODGE', 'CHRY', 'VOLVO',
                'TRIUMPH', 'SCANIA', 'CRONOS', 'KIA', 'MERCEDES ', 'LR', 'JAC', 'PORSCHE', 'LEXUS', 'MITSUBISHI', 'SUBARU',
               'FERRARI', 'MINI', 'MMC', 'MASERATI', 'LIFAN']
for i in lista_marcas:
    dividir_marca_veiculo(i)
```

É comum encontrarmos diferentes nomes que se referem à mesma marca. Isso pode ser devido a erros de digitação ao registrar as informações ou mesmo ao fato de que uma mesma marca pode ser conhecida por nomes diferentes. Para lidar com essa questão, é necessário corrigir os dados, padronizando os nomes para facilitar a visualização.
